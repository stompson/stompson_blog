```python
data_dir='/Users/steventompson/Dropbox/Personal Stuff/CFB_stats'
home_dir='/Users/steventompson/Git/stompson_blog'
data_dir=opj(home_dir,'data')
fig_dir=opj(home_dir,'figures')
```

```python
team_dict={}
topteam_dict={}
years=[2014,2015,2016,2017,2018]
for year in years:
    sp=pd.read_csv(opj(data_dir,'sp+_ratings','sp+_ratings_{}.csv'.format(year)), encoding='utf-8')
    topteam_dict[year]=sp.loc[range(25),'team']
    team_dict[year]=sp
    
```

```python
game_dict={}
for year in years:
    for team in topteam_dict[year]:
        game_dict['{}_{}'.format(team,year)]=cfb.get_game_info(year=year,team=team,seasonType='both')
```

```python
colvars=['rating','offenserating','defenserating',
         'offensesuccess', 'offenseexplosiveness','offenserushing','offensepassing',
         'offensestandardDowns','offensepassingDowns','offenserunRate','offensepace',
         'defensesuccess', 'defenseexplosiveness','defenserushing','defensepassing',
         'defensestandardDowns','defensepassingDowns',
         'defensehavoctotal','defensehavocfrontSeven','defensehavocdb','specialTeamsrating']

t1vars=['t1_'+var for var in colvars]
t2vars=['t2_'+var for var in colvars]
newvars=['year','team1','team2','t1_score','t2_score','home_team',
         't1_off_exp_pts','t1_def_exp_pts','t2_off_exp_pts','t2_def_exp_pts']+t1vars+t2vars

def get_team_sp_opponent_stats(df1,team,year):
    newdf1=pd.DataFrame(columns=newvars)
    for ix in df1.index:
        home=df1.loc[ix,'home_team']
        away=df1.loc[ix,'away_team']
        newdf1.loc[ix,'year']=year
        newdf1.loc[ix,'team1']=team
        newdf1.loc[ix,'home_team']=home
        off_home_points=[hp for ix,hp in enumerate(df1['home_points']) if df1.loc[ix,'home_team']==team]
        off_away_points=[hp for ix,hp in enumerate(df1['away_points']) if df1.loc[ix,'away_team']==team]
        def_home_points=[hp for ix,hp in enumerate(df1['home_points']) if df1.loc[ix,'home_team']!=team]
        def_away_points=[hp for ix,hp in enumerate(df1['away_points']) if df1.loc[ix,'away_team']!=team]
        newdf1.loc[ix,'t1_off_avg_pts']=np.mean(off_home_points+off_away_points)
        newdf1.loc[ix,'t1_def_avg_pts']=np.mean(def_home_points+def_away_points)

        if home==team:
            newdf1.loc[ix,'t1_score']=df1.loc[ix,'home_points']    
            newdf1.loc[ix,'t1_off_exp_pts']=team_dict[year].loc[team_dict[year]['team']==home,'offenserating'].values[0]
            newdf1.loc[ix,'t1_def_exp_pts']=team_dict[year].loc[team_dict[year]['team']==home,'defenserating'].values[0]
            newdf1.loc[ix,t1vars]=team_dict[year].rank(ascending=False).loc[team_dict[year]['team']==home,colvars].values

            newdf1.loc[ix,'team2']=away
            newdf1.loc[ix,'t2_score']=df1.loc[ix,'away_points']
            #newdf1.loc[ix,'t2_off_exp_pts']=team_dict[year].loc[team_dict[year]['team']==away,'offenserating'].values[0]
            #newdf1.loc[ix,'t2_def_exp_pts']=team_dict[year].loc[team_dict[year]['team']==away,'defenserating'].values[0]
            newdf1.loc[ix,t2vars]=team_dict[year].rank(ascending=False).loc[team_dict[year]['team']==away,colvars].values
        else:
            newdf1.loc[ix,'t1_score']=df1.loc[ix,'away_points']    
            newdf1.loc[ix,'t1_off_exp_pts']=team_dict[year].loc[team_dict[year]['team']==away,'offenserating'].values[0]
            newdf1.loc[ix,'t1_def_exp_pts']=team_dict[year].loc[team_dict[year]['team']==away,'defenserating'].values[0]
            newdf1.loc[ix,t1vars]=team_dict[year].rank(ascending=False).loc[team_dict[year]['team']==away,colvars].values

            newdf1.loc[ix,'team2']=home
            newdf1.loc[ix,'t2_score']=df1.loc[ix,'home_points']
            #newdf1.loc[ix,'t2_off_exp_pts']=team_dict[year].loc[team_dict[year]['team']==home,'offenserating'].values[0]
            #newdf1.loc[ix,'t2_def_exp_pts']=team_dict[year].loc[team_dict[year]['team']==home,'defenserating'].values[0]
            newdf1.loc[ix,t2vars]=team_dict[year].rank(ascending=False).loc[team_dict[year]['team']==home,colvars].values
    return newdf1
```

```python
game_df=pd.DataFrame(columns=newvars)
for key in game_dict.keys():
    tm=key.split('_')[0]
    yr=int(key.split('_')[1])
    tm_df=game_dict[key]
    new_tm_df=get_team_sp_opponent_stats(df1=tm_df,team=tm,year=yr)
    game_df=game_df.append(new_tm_df,sort=False)

game_df=game_df.reset_index()
fbs_ixes=[ix for ix,val in enumerate(game_df['t2_rating']) if len(val)>=1]
game_df=game_df.loc[fbs_ixes,:].reset_index()

for var in t1vars+t2vars:
    for ix,val in enumerate(game_df[var]):
        if isinstance(val,list):
            game_df.loc[ix,var]=val[0]

game_df['win']=0
win_ixes=game_df['t1_score']>game_df['t2_score']
game_df.loc[win_ixes,'win']=1
```
